```python
from sklearn.preprocessing import MinMaxScaler
from keras.models import Sequential
from keras.layers import Dense, LSTM
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = pd.read_csv('train_1.csv').fillna(0)
```

```python
data
```

```python
data = pd.melt(data[list(data.columns)], id_vars= 'Page', var_name= 'date', value_name= 'visit')
data
```

```python
data['date'] = data['date'].astype('datetime64[ns]')
data['date']
```

```python
# This gets us a list of all unique page name
un = data['Page'].unique()
un
```

```python
#Random page name
un[3]
```

**Grouping data**

```python
gp = data.groupby(['Page'])
```

```python
ser = gp.get_group(un[3])['visit']
ser = np.array(ser)
ser = ser.reshape(-1,1)
scaler = MinMaxScaler(feature_range=(0,1))
Normalized = scaler.fit(ser)
Normalized = scaler.fit_transform(ser)
Normalized
```

**LSTM MODEL**

```python
model = Sequential()

model.add(LSTM(20, activation = 'relu', return_sequences= True, input_shape = (None, 1)))
model.add(LSTM(20, activation = 'relu'))
model.add(Dense(1))
model.compile(optimizer = 'adam', loss = 'mse', metrics = ['accuracy'])
```
